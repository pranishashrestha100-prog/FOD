```python
import pandas as pd
```

```python
df=pd.read_csv('Health_heart_experimental.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.lineplot(data=df,x='age',y='HR',marker='.',label='HR')
sns.lineplot(data=df,x='age',y='BMI',marker='.',label='BMI')
plt.title('LINE GRAPH')
plt.xlabel('age')
plt.ylabel('HR,BMI')
plt.show()
```

Description
This code is used to draw two line graphs. One line shows how heart rate (HR) changes with age and the other shows how BMI changes with age. Markers show individual points. The title and axis labels(x and y axis) explain the graph and plt.show() displays it.



```python
df['sex'] = df['sex'].replace({0:'Female',1:'Male'})
plt.figure()
df['sex'].value_counts().plot(kind='bar')
plt.xlabel("Sex")
plt.ylabel("Number")
plt.title("Sex Distribution")
plt.show()
```

In this bar plot,no of males and females are counted. The title is sex distribution and total no of males and females are counted. This code replaces 0 as female and 1 as male. The x-axis is labeled as sex and y axis as number.

```python
df['sex'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=90,colors=['pink','skyblue'])
plt.title('Sex Distribution')
plt.ylabel('')
plt.show()
```

The pie chart to show how many males and females are in the data. It shows each groupâ€™s percentage, adds a title, uses pink for female and sky blue color for male ,and then displays the chart.Autopct='%1.1f%%' shows percentage with one decimal, and startangle=90 rotates the pie chart clockwise.

```python
plt.boxplot([df['weightKg'],df['heightCm']])
plt.xticks([1,2],['weightKg','heightCm'])
plt.title('Box plot for weight and height')
plt.show()
```

This code makes a box plot to show the spread of weightKg and heightCm values. Each box shows the minimum, maximum, middle value (median), and any outliers. The x-axis labels the boxes, a title is added, and plt.show() displays the chart. It helps compare weight and height visually.


```python
plt.figure()
plt.hist(df['weightKg'],bins=30)
plt.xlabel("weightKg")
plt.ylabel("Frequency")
plt.title("Weight Distribution")
plt.show()
```

This histogram shows how weight values are distributed. It divides the data into 30 bins, counts how many values fall into each bin, labels the x-axis as weightKg and y-axis as frequency, adds a title and plt.show() displays the chart.

```python
df = pd.read_csv('Health_heart_experimental.csv').sample(200)
plt.scatter(df['age'], df['BMI'],color='g',marker='*')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Age vs BMI')
plt.show()
```

It reads a health dataset, randomly selects 200 records, and creates a scatter plot of Age versus BMI. Each dot represents one person. Green dots are used to show how BMI changes with age. Labels are added to both axes, a title is given, and the plot is displayed.
